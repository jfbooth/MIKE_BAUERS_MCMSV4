```python
import numpy as np 
import xarray as xr
import netCDF4
import matplotlib.pyplot as plt

import os 
import glob
```

```python
year_list = [2008, 2012]

folder = '/mnt/drive5/ERAINTERIM'

for year in range(year_list[0], year_list[1]+1):
    u_file = os.path.join(folder, f'U10/U10_{year}.nc') 
    v_file = os.path.join(folder, f'V10/V10_{year}.nc') 
    
    u_ds = xr.open_dataset(u_file)
    v_ds = xr.open_dataset(v_file)
    u = u_ds['var165']
    v = v_ds['var166']
    u_ds.close()
    v_ds.close()
    
    w = (u**2 + v**2)**.5
    
    # creating the output file
    out_ds = xr.Dataset({'uv10': w})
   
    print(year)
    break
    
print(u_ds.keys())
   
    
```
